```python
%matplotlib inline
import pyImageEditor as imgEdit

import matplotlib.pyplot as plt
from IPython.display import Image
```

## Image that we will be working with

```python
myFilter = imgEdit.pyfilter("./img/arda.jpg") # Initialize pyfilter class

fig, ax = plt.subplots(figsize=(9, 9))
ax.imshow(myFilter.getImage())
plt.show()
```

## Using the Gaussian Smoothing Function

```python
originalImage = myFilter.getImage() # Get original image from pyfilter object
gaussianFilteredImage = myFilter.gaussianSmooth() # Apply gaussian smoothing function

print("Original Image:")
fig, ax = plt.subplots(figsize=(9, 9))
ax.imshow(originalImage)
plt.show()
print("Gaussian Smoothed Image:")
fig, ax = plt.subplots(figsize=(9, 9))
ax.imshow(gaussianFilteredImage)
plt.show()
```

## Using the Mean Filtering Function

```python
originalImage = myFilter.getImage() # Get original image from pyfilter object
meanFilteredImage = myFilter.meanFilter() # Apply mean filtering function

print("Original Image:")
fig, ax = plt.subplots(figsize=(9, 9))
ax.imshow(originalImage)
plt.show()
print("Mean Filtered Image:")
fig, ax = plt.subplots(figsize=(9, 9))
ax.imshow(meanFilteredImage)
plt.show()
```

## Comparing the Gaussian Filter and Mean Filter

```python
print("Gaussian Filtered Image:")
fig, ax = plt.subplots(figsize=(9, 9))
ax.imshow(gaussianFilteredImage)
plt.show()
print("Mean Filtered Image:")
fig, ax = plt.subplots(figsize=(9, 9))
ax.imshow(meanFilteredImage)
plt.show()
```
